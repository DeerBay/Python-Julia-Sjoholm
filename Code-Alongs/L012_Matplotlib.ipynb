# Matplotlib

```python
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y = [value**2 for value in x]

y2 = [value**3 for value in x]

print(f"{x = }")
print(f"{y = }")

plt.plot(x, y, label= "x**2")
plt.plot(x, y2, label= "x**3")
plt.legend() #Fö att namge linjerna, se ovan
```

```python
x = [0, -100, 25, 65, -323]
y = [0, 3, 7, 3, 9]

plt.title("My graph")
plt.xlabel("x-values")
plt.ylabel("y-values")

# plt.axis([-50, 80, 2, 8]) används för att zooma in på vissa värden
plt.plot(x, y)
plt.grid() #för rutnät (kan bestämma avtånd på "gridds")
plt.plot(x, y, color="red", linestyle="dashed", marker="o")
```

## Scatter

```python
x = list(range(-10, 11))
y = [value**2 for value in x]

plt.scatter(x, y, marker="x")
```

## Bars

```python
labels = ["Male", "Female"]
values = [15, 23]

plt.bar(labels, values)
```

## Pie Charts

```python
plt.pie(values, labels=labels)
```

## Sub plots

```python
x = list(range(-10, 11))
y = [value**2 for value in x]
y2 = [value**3 for value in x]

fig, axs = plt.subplots(1, 2)
axs[0].plot(x, y)
axs[1].plot(x, y2)
```
